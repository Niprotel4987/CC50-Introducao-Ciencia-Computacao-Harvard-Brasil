```
import sympy as sp
x = sp.symbols('x')
sinc = sp.Piecewise(
    (1, x == 0),
    (sp.sin(x)/x, True)
)

display(sp.Eq(sp.Function("sinc")(x), sinc))
x = sp.symbols('x')
```

```
import sympy as sp
from IPython.display import display

x1, x2, x3, n = sp.symbols('x1 x2 x3 n')
mean = (x1 + x2 + x3) / 3

display(sp.Eq(sp.Symbol("mean"), mean))
```

```
import sympy as sp
from IPython.display import display

x = sp.symbols('x')

f = x**3 + 2*x**2 - x + 1
derivada = sp.diff(f, x)

display(sp.Eq(sp.Symbol("f(x)"), f))
display(sp.Eq(sp.Symbol("f'(x)"), derivada))
```

```
integral = sp.integrate(f, x)
display(sp.Eq(sp.Symbol("∫f(x)dx"), integral))

```

```
equacao = x**2 - 5*x + 6
solucao = sp.solve(equacao, x)

display(solucao)
```

```
limite = sp.limit(sp.sin(x)/x, x, 0)
display(limite)
```

```
f_piece = sp.Piecewise(
    (x**2, x < 0),
    (x, x >= 0)
)

display(sp.Eq(sp.Symbol("f(x)"), f_piece))
```

```
import matplotlib.pyplot as plt

# Rótulos dos produtos
rotulos = ['Produto A', 'Produto B', 'Produto C', 'Produto D']

# Valores percentuais
valores = [45, 25, 20, 10]

# Criando a figura e o eixo
figura, eixo = plt.subplots()

# Criando o gráfico de rosca
eixo.pie(
    valores,
    labels=rotulos,
    autopct='%1.1f%%',   # Mostra porcentagem com 1 casa decimal
    startangle=90,      # Inicia o gráfico a partir de 90 graus
    wedgeprops=dict(width=0.4)  # Define a largura da rosca
)

# Título do gráfico
eixo.set_title('Distribuição de Receita')

# Exibindo o gráfico
plt.show()
```

```
import matplotlib.pyplot as plt

# Rótulos dos produtos
rotulos = ['Produto A', 'Produto B', 'Produto C', 'Produto D']

# Valores percentuais
valores = [50, 20, 15, 15]

# Define qual fatia será destacada
destaque = (0.1, 0, 0, 0)  # Apenas o primeiro produto será afastado

# Criando o gráfico de pizza destacada
plt.pie(
    valores,
    labels=rotulos,
    explode=destaque,
    autopct='%1.1f%%',
    startangle=90
)

# Título do gráfico
plt.title('Gráfico de Pizza Destacada – Distribuição de Vendas')

# Garante que o gráfico fique circular
plt.axis('equal')

# Exibe o gráfico
plt.show()
```

```
import matplotlib.pyplot as plt

# Lista de meses
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']

# Vendas dos produtos
produto_a = [30, 40, 45, 50, 60, 70]
produto_b = [20, 30, 35, 40, 45, 50]
produto_c = [10, 20, 25, 30, 35, 40]

# Criando o gráfico de área empilhada
plt.stackplot(meses, produto_a, produto_b, produto_c)

# Rótulos dos eixos
plt.xlabel("Meses")
plt.ylabel("Vendas")

# Título do gráfico
plt.title("Exemplo de Gráfico de Área Empilhada")

# Legenda
plt.legend(["Produto A", "Produto B", "Produto C"])

# Exibindo o gráfico
plt.show()
```

```
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

ax.set_xlim(0, 10)
ax.set_ylim(0, 5)

# Corpo
corpo = plt.Circle((2, 2), 1, color='#8B4513')
ax.add_patch(corpo)

# Cabeça
cabeca = plt.Circle((3, 2.5), 0.5, color='#8B4513')
ax.add_patch(cabeca)

# Orelhas
orelha1 = plt.Circle((3.2, 3), 0.15, color='#654321')
orelha2 = plt.Circle((2.8, 3), 0.15, color='#654321')
ax.add_patch(orelha1)
ax.add_patch(orelha2)

# Olho
olho = plt.Circle((3.2, 2.6), 0.05, color='black')
ax.add_patch(olho)

ax.set_aspect('equal')
ax.axis('off')

def animar(frame):
    x = 2 + frame * 0.05
    corpo.center = (x, 2)
    cabeca.center = (x + 1, 2.5)
    orelha1.center = (x + 1.2, 3)
    orelha2.center = (x + 0.8, 3)
    olho.center = (x + 1.2, 2.6)
    return corpo, cabeca, orelha1, orelha2, olho

ani = animation.FuncAnimation(
    fig, animar, frames=120, interval=50, blit=True
)

# 🔥 Salvar como GIF
ani.save("capivara_animada.gif", writer="pillow", fps=20)

print("GIF salvo como capivara_animada.gif")
```
